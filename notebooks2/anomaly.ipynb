```python
from matplotlib import pyplot as plt
from scipy import stats
import seaborn as sns
```

```python
# Load some dataset (e.g., Titanic) and experiment with these algorithms

# load dataset titanic
df = sns.load_dataset('titanic')

# dimensions of your dataframe
print('Dimensions:', df.shape[0], 'rows and ', df.shape[1], 'columns')

# Data types of your dataframe
print(df.dtypes)
```

```python
# Boxplot can help you visualise potential outliers. Take any numerical attribute and plot it along your class
# What do we see in a boxplot? are there any outliers? (univariate approach)

plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='class', y='fare', hue='survived')
plt.ylim(0, 300)  # Limiting y-axis to 300 for better visualization
plt.title('Fare distribution by Passenger Class and Survival')
plt.show()
```

```python
# do you see any potential outliers in fare attribute? (univariate approach)
df['fare'].hist(bins=30)
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Distribution of Fare')
plt.show()
```

```python
# Sanity check: Q-Q plot to check if fare deviates from a normal distribution (univariate approach)

stats.probplot(df['fare'].dropna(), dist="norm", plot=plt)
plt.title("Q-Q Plot of Regression Residuals")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Residuals")
plt.show()
```
